```python
# unit step
import numpy as np
import matplotlib.pyplot as plt

def unit_step(t):
    return np.heaviside(t, 1)

t = np.linspace(-50, 50, 1000)
u = unit_step(t)
plt.plot(t, u)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Unit Step Function')
plt.grid(True)
plt.show()
```

```python
# unit ramp
import numpy as np
import matplotlib.pyplot as plt

def unit_ramp(t):
    return np.maximum(t, 0)

t = np.linspace(-5, 5, 1000)
ramp = unit_ramp(t)
plt.plot(t, ramp)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Unit Ramp Function')
plt.grid(True)
plt.show()
```

```python
# sine function

import numpy as np
import matplotlib.pyplot as plt

amplitude = 40  
frequency = 8  
phase = 0.0     


fs = 5800  
t = np.arange(0, 3, 1/fs)

sin_wave = amplitude * np.sin(2*np.pi*frequency*t + phase)
plt.plot(t, sin_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.grid(True)
plt.show()

```

```python
# cosine wave

import numpy as np
import matplotlib.pyplot as plt

amplitude = 40  
frequency = 8  
phase = np.pi/2  

fs = 5800  
t = np.arange(0, 1, 1/fs)
cos_wave = amplitude * np.cos(2*np.pi*frequency*t + phase)
plt.plot(t, cos_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Cosine Wave')
plt.grid(True)
plt.show()
```

```python
# sawtooth wave


import numpy as np
from scipy import signal
import matplotlib.pyplot as plot
timePoints = np.linspace(0, 1, 500)
plot.plot(timePoints, signal.sawtooth(2 * np.pi * 8 * timePoints))
plot.title('Sqaure wave - 8 Hz sampled at 1000 Hz /second')
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.ylim(-2, 2)
plot.show()
```

```python
# square signal

import numpy as np
import matplotlib.pyplot as plt

amplitude = 40  
frequency = 8  


fs = 5800  
t = np.arange(0, 1, 1/fs)

square_wave = amplitude * signal.square(2 * np.pi * frequency * t)
plt.plot(t, square_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Square Wave')
plt.ylim(-50, 50)  
plt.grid(True)
plt.show()
```

```python
# triangle signal

from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

amplitude = 40  
frequency = 8  

fs = 100 
t = np.arange(0, 1, 1/fs)
triangle_wave = amplitude * signal.sawtooth(2 * np.pi * frequency * t, width=0.5)
plt.plot(t, triangle_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Triangle Wave')
plt.ylim(-50, 50)  
plt.grid(True)
plt.show()
```

```python
# exponential increasing wave

import numpy as np
import matplotlib.pyplot as plt

amplitude = 40  
decay_factor = 0.1  


fs = 5800  
t = np.arange(0, 1, 1/fs)
exponential_wave = amplitude * np.exp(decay_factor * t)
plt.plot(t, exponential_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Exponential Increasing Wave')
plt.grid(True)
plt.show()
```

```python
# exponential decreasing wave

import numpy as np
import matplotlib.pyplot as plt

amplitude = 40  
decay_factor = 0.1  
fs = 5800 
t = np.arange(0, 1, 1/fs)
exponential_wave = amplitude * np.exp(-decay_factor * t)

plt.plot(t, exponential_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Exponential Decreasing Wave')
plt.grid(True)
plt.show()
```

```python
# gaussian signal

import numpy as np
import matplotlib.pyplot as plt

amplitude = 40 
mean = 0.0  
std_dev = 0.5  
fs = 5800  
t = np.arange(-5, 5, 1/fs)
gaussian_signal = amplitude * np.exp(-(t - mean)**2 / (2 * std_dev**2))

plt.plot(t, gaussian_signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Gaussian Signal')
plt.grid(True)
plt.show()
```

```python
# sinc signal

import numpy as np
import matplotlib.pyplot as plt

amplitude = 40  
frequency = 8  
fs = 5800  
t = np.arange(-5, 5, 1/fs)
sinc_signal = amplitude * np.sinc(frequency * t)

plt.plot(t, sinc_signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sinc Signal')
plt.grid(True)
plt.show()
```

```python
# addition of two signals

import numpy as np
import matplotlib.pyplot as plt

fs = 1000  
t = np.arange(0, 1, 1/fs)  
f1 = 50  
signal1 = np.sin(2*np.pi*f1*t) 

f2 = 120  
signal2 = 0.5 * np.sin(2*np.pi*f2*t)  

sum_signal = signal1 + signal2

plt.figure(figsize=(8, 4))
plt.plot(t, signal1, label='Signal 1')
plt.plot(t, signal2, label='Signal 2')
plt.plot(t, sum_signal, label='Sum of Signals')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal Addition')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# subtraction of two signals

import numpy as np
import matplotlib.pyplot as plt

fs = 1000 
t = np.arange(0, 1, 1/fs)  
f1 = 50  
signal1 = np.sin(2*np.pi*f1*t) 

f2 = 120  
signal2 = 0.5 * np.sin(2*np.pi*f2*t) 

diff_signal = signal1 - signal2

plt.figure(figsize=(8, 4))
plt.plot(t, signal1, label='Signal 1')
plt.plot(t, signal2, label='Signal 2')
plt.plot(t, diff_signal, label='Difference of Signals')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal Subtraction')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Multiplication of two signals in python

import numpy as np
import matplotlib.pyplot as plt

fs = 1000  
t = np.arange(0, 1, 1/fs)  
f1 = 50  
signal1 = np.sin(2*np.pi*f1*t)  

f2 = 120  
signal2 = 0.5 * np.sin(2*np.pi*f2*t)  

mult_signal = signal1 * signal2
plt.figure(figsize=(8, 4))
plt.plot(t, signal1, label='Signal 1')
plt.plot(t, signal2, label='Signal 2')
plt.plot(t, mult_signal, label='Multiplication of Signals')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal Multiplication')
plt.legend()
plt.grid(True)
plt.show()
```
